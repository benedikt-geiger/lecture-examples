```python
import numpy as np 
import scipy as sc 
import matplotlib.pyplot as plt
```

```python
seed=2024
sequence_length=int(1e4)

rng = np.random.default_rng(seed)

white_noise_sequence=rng.standard_normal(sequence_length)


```

```python
plt.plot(white_noise_sequence)
```

```python
plt.plot(np.correlate(white_noise_sequence,white_noise_sequence,'same'))
```

```python
signal_filter=np.ones(shape=(128))

signal=np.convolve(white_noise_sequence,signal_filter)
```

```python
plt.plot(signal)
```

```python
plt.plot(np.correlate(signal,signal,'same'))
```
